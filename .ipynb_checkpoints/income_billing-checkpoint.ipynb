```python
import pandas as pd
import numpy as np
```

### Load Brands By Distributor

```python
brands = 'data/BrandLinesByDistributors.xlsx'
brands_df = pd.read_excel(brands)
```

```python
brands_df.head()
```

### Remove Unnamed Columns

```python
for col in brands_df.columns:
    if 'Unnamed' in col:
        del brands_df[col]
```

### Melt Columns

```python
brands_df = pd.melt(brands_df)
brands_group = brands_df.groupby(['variable', 'value']).count().reset_index()
```

```python
brands_group = brands_group[brands_group['value'] != 'Total Lines']
```

```python
rawdata = "data/oldMarketReport.xlsx"
rawdata_df = pd.read_excel(rawdata)
```

```python
rawdata_df.head()
```

```python
data = rawdata_df.groupby(['distributor','client', 'tapname','city', 'zonevalue'])['quantity'].sum().reset_index()
```

```python
data.head(30)
```

```python
unique_locations = data.client.unique()

location_obj = {}
for x in unique_locations:
    loc = x
    df = data[data['client'] == loc]
    post = {'zone': df.zonevalue.unique()[0],
            'city': df.city.unique()[0],
           'distributors': df.distributor.unique(),
           'brands': df.tapname.unique()}
    location_obj[x] = post
location_obj
```

```python
market_file = 'data/MarketReport.xlsx'
market_df = pd.read_excel(market_file)
market_df.head()
```

### DISTRIBUTOR CHECK

```python
market_unique_loc = market_df.client.unique()
i = 0

error_check = {}

for x in market_unique_loc:
    loc = x
    df = market_df[market_df['client'] == x]
    for index, row in df.iterrows():
        zone = row['zonevalue']
        try:
            comparison_zone = location_obj[x]['zone']
            if zone != comparison_zone:
                print(f'client {x} not associated with {zone}')
        except:
            pass
        distributor = row['distributor']
        try:
            comparison_dist = location_obj[x]['distributors']
            if distributor not in comparison_dist:
                i = i + 1
                print(i)

        
#                 print(f'client {x} not associated with {distributor}')
        except:
            pass
```
